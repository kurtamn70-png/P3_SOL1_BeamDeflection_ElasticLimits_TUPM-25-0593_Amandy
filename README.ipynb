# SOL-01: Wing Spar Beam Deflection & Elastic Limits

**Mechanical Engineering - Computer Programming and Fundamentals Final Project**

## Project Overview

This project applies Engineering Data Analytics to analyze the structural integrity of a **Synthetic Wing Spar Digital Twin**. By processing real-time telemetry data, the pipeline calculates beam deflection under varying load factors (G-forces) to ensure the component remains within its **linear-elastic limits** and meets safety factors for flight.

## Features

* **Data Pipeline:** Automated cleaning, duplicate removal, and node-specific filtering (Node 19) of structural datasets.
* **Engineering Analytics:** Descriptive statistics (Mean, Median, Std Dev, Variance) and high-precision correlation analysis for Hooke's Law verification.
* **Visualization:** 3 Static Engineering Charts and an Animated Simulation of structural response dynamics.
